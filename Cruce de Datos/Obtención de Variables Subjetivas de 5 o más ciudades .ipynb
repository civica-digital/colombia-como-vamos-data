```python
import pandas as pd
import numpy as np
import csv
import os
```

```python
variable_dict = {}
file_dict = {}
for fn in os.listdir('diccionarios'):
    stringfn = 'diccionarios/' + fn
    data = pd.read_csv(stringfn,delimiter=",", dtype=np.string_, na_values=[" "], encoding="utf-8")
    list_cnames = list(data.variable)
    for name in list_cnames:
        if name in variable_dict:
            variable_dict[name] = variable_dict[name] + 1
            file_dict[name].append(fn)
        else:
            variable_dict[name] = 1
            file_dict[name] = []
```

```python
a = {}
a['b'] = 2
a['c'] = 3
```

```python
list_keys_plus5 = []
for key in variable_dict:
    if variable_dict[key] >= 5:
        list_keys_plus5.append(key)
```

```python
import json
with open('variables_subjetivas.json', 'w') as fp:
    json.dump(variable_dict, fp)
```

```python
for fn in os.listdir('datos'):
    stringfn = 'datos/' + fn
    outfn = 'dd_producto/' + fn
    print(stringfn)
    data = pd.read_csv(stringfn,delimiter=",", dtype=np.string_, na_values=[" "], encoding="utf-8")
    columnas_subset= set(data.columns).intersection(list_keys_plus5)
    datos_filtrados = data.filter(columnas_subset)
    datos_filtrados.to_csv(outfn,quoting=csv.QUOTE_NONNUMERIC, encoding="utf-8", na_rep = np.nan, index = False)
```

```python
for fn in os.listdir('diccionarios'):
    stringfn = 'diccionarios/' + fn
    outfn = 'dd_producto/' + fn
    print(stringfn)
    dict_data = pd.read_csv(stringfn,delimiter=",", dtype=np.string_, na_values=[" "], encoding="utf-8")
    dictionary_filtered = dict_data.loc[dict_data['variable'].isin(list_keys_plus5)]
    dictionary_filtered.to_csv(outfn,quoting=csv.QUOTE_NONNUMERIC, encoding="utf-8", na_rep = np.nan, index = False)

    
    
```
