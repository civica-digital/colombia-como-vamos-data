```python
import pandas as pd
import numpy as np
import csv
```

```python
datos = pd.read_csv("objetivos_datos-csv.csv",delimiter=",", dtype=np.string_, na_values=[" "], encoding="utf-8")
```

```python
diccionario = pd.read_csv("objetivos_diccionario-csv.csv",delimiter=",", dtype=np.string_, na_values=[" "], encoding="utf-8")
```

```python
for ciudad in np.unique(datos.CIUDAD):
    filename = "datos/datos_objetivos_"+ ciudad + ".csv"
    filename_diccionario = "datos/diccionario_datos_objetivos_"+ ciudad +".csv"
    datos_ciudad = datos[datos['CIUDAD'] == ciudad]
    datos_ciudad_sinna = datos_ciudad.dropna(axis=1,how="all")
    datos_ciudad_sinna.to_csv(filename,quoting=csv.QUOTE_NONNUMERIC, encoding="utf-8", na_rep = np.nan, index = False)
    columnas_ciudad = datos_ciudad_sinna.columns.values
    diccionario_local = diccionario.loc[diccionario['id'].isin(columnas_ciudad)]
    diccionario_local.to_csv(filename_diccionario,quoting=csv.QUOTE_NONNUMERIC, encoding="utf-8", na_rep = np.nan, index = False)
```
